```python
Table1 = pd.DataFrame({'custID':['1001','1002','1003','1004'],'Q1':['101','102','103','104'],'Q2':['201','202','203','204']})
Table2 = pd.DataFrame({'custID':['1001','1006','1003','1004'],'Q3':['301','302','303','304'],'Q4':['401','402','403','404']})
```

```python
Table1
```

```python
Table2
```

# merging

```python
pd.merge(Table1,Table2,how='outer',on='custID')
```

```python
pd.merge(Table1,Table2,how='inner',on='custID')
```
